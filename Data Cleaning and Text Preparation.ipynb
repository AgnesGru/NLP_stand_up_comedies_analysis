```python
import nltk
from nltk.stem.porter import PorterStemmer
from nltk.corpus import stopwords
```

```python
# Pandas DataFrame
import pandas as pd

data_df = pd.DataFrame.from_dict(data_combined, orient='index', columns = ['transcript']) # to to samo co '.transpose'
data_df
```

```python
# data_df.transcript.loc['ali']
```

```python
import re 
import string
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

def clean_text1(text):    
    text = text.lower()
    text = re.sub(r'[\?,.*]*', '', text) # replacement of what in the square bracket
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text) # remove all punctuation
    text = re.sub('\w*\d\w*', '', text)# remove digits
    text = re.sub('[‘’“”…]', '', text)
    text = re.sub('\n', '', text)
    text = re.sub(re.compile(r'\W*\b\w{1,3}\b'), "", text) #remove short words
    text = text.split()
    text = [word for word in text if not word in stopwords.words('english')]
    words = [lemmatizer.lemmatize(word) for word in text if not word in stopwords.words('english')]
    text = ' '.join(words)
    return text

round1 = lambda x: clean_text1(x)
```

```python
data_clean = pd.DataFrame(data_df.transcript.apply(round1))
data_clean
```

### Organising the data

```python
# Corpus - a collection of text
# data_df

full_names = ['Ali Wong', 'Anthony Jeselnik', 'Bill Burr', 'Bo Burnham', 'Dave Chappelle', 'Hasan Minhaj',
              'Jim Jefferies', 'Joe Rogan', 'John Mulaney', 'Louis C.K.', 'Mike Birbiglia', 'Ricky Gervais']

data_clean['full_name'] = full_names
data_clean
```

```python
# pickle it for later use
data_clean.to_pickle("corpus.pkl")
```

```python
# Document-Term Matrix - word counts in matrix format
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer()
data_cv = cv.fit_transform(data_clean.transcript)
data_dtm = pd.DataFrame(data_cv.toarray(), columns=cv.get_feature_names())
data_dtm.index = data_clean.index
data_dtm
```

```python
# Let's pickle it for later use
data_dtm.to_pickle("dtm.pkl")
```
